```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
x=np.linspace(-3,3,1000)
y=[]

def gauss(x,mu=0,s=1):
    g=1/(s*(2*np.pi)**0.5)*np.exp(-0.5*(x-mu)**2/(s**2))
    return g

for i in x:
    y.append(gauss(i))
y=np.array(y)
y=y-0.5
y=y*3
plt.title("Normal Distribution")
plt.plot(x,y)
plt.show()
```
